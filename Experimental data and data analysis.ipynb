```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Raw data:
## PFUs from free phages in the supernatant under 2 different conditions (Glucose or Arsenate and Azide) and 3 types of bacteria (Permissive, Resistant, Buffer)

```python
# Data groups for each phage
data_groups = {
    "λ_GLU": {
        "Perm": [11, 9, 10],
        "Res": [1500, 1000, 800],
        "Buf": [1100, 1100, 900]
    },
    "λ_AsAz": {
        "Perm": [150, 140, 90],
        "Res": [1800, 1300, 700],
        "Buf": [1700, 1300, 1400]
    },
    "φ80_GLU": {
        "Perm": [3400, 2000, 2000],
        "Res": [16000, 10000, 8000],
        "Buf": [13000, 10000, 11000]
    },
    "φ80_AsAz": {
        "Perm": [7000, 5700, 5000],
        "Res": [14000, 15000, 9000],
        "Buf": [11000, 10000, 10000]
    },
    "m13_GLU": {
        "Perm": [10000, 10000, 13000],
        "Res": [28000, 28000, 35000],
        "Buf": [30000, 38000, 40000]
    },
    "m13_AsAz": {
        "Perm": [33000, 33000, 32000],
        "Res": [36000, 33000, 36000],
        "Buf": [30000, 35000, 41000]
    },
    "T6_GLU": {
        "Perm": [27000, 31000, 18000],
        "Res": [37000, 41000, 36000],
        "Buf": [45000, 50000, 33000]
    },
    "T6_AsAz": {
        "Perm": [41000, 40000, 37000],
        "Res": [38000, 32000, 36000],
        "Buf": [48000, 38000, 38000]
    },
    "T5_GLU": {
        "Perm": [400, 800, 200],
        "Res": [17000, 10000, 4000],
        "Buf": [16000, 12000, 12000]
    },
    "T5_AsAz": {
        "Perm": [400, 1000, 200],
        "Res": [17000, 13000, 4000],
        "Buf": [15000, 18000, 14000]
    }
}
```

```python
# Function to calculate mean, standard deviation, and standard error of the mean
def calculate_stats(data):
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # Sample standard deviation (n-1)
    sem = std / np.sqrt(len(data))  # Standard error of the mean
    return mean, std, sem
```

```python
# Calculate and store results
results = []
for phage, groups in data_groups.items():
    for group_name, data in groups.items():
        mean, std, sem = calculate_stats(data)
        results.append({
            "Phage": phage,
            "Group": group_name,
            "Exp 1": data[0],
            "Exp 2": data[1],
            "Exp 3": data[2],
            "Mean": round(mean, 0),
            "STD": round(std, 0),
            "SEM": round(sem, 0)
        })

# Convert results into a pandas DataFrame
df = pd.DataFrame(results)

# Display the results
df
```

## Plot the average number of PFUs

```python
# We want to plot 5 phages: λ, φ80, m13, T6, and T5.
# For each phage, we combine the two conditions (GLU and As/Az) for each of the 3 groups:
# Order: [GLU-Perm, AsAz-Perm, GLU-Res, AsAz-Res, GLU-Buf, AsAz-Buf]
phages = ["λ", "φ80", "m13", "T6", "T5"]
conditions = ["Perm", "Res", "Buf"]

avg_free_phage = []  # This will be a list of 5 lists, each with 6 numbers
sem_free_phage = []  # Same structure, containing SEM values

for ph in phages:
    sublist_avg = []
    sublist_sem = []
    for cond in conditions:
        # Filter the DataFrame for the GLU condition for this phage/condition
        row_glu = df[(df["Phage"] == f"{ph}_GLU") & (df["Group"] == cond)]
        row_asaz = df[(df["Phage"] == f"{ph}_AsAz") & (df["Group"] == cond)]
        # Extract the Mean and SEM values (assume one row per combination)
        avg_glu = row_glu["Mean"].values[0]
        avg_asaz = row_asaz["Mean"].values[0]
        sem_glu = row_glu["SEM"].values[0]
        sem_asaz = row_asaz["SEM"].values[0]
        sublist_avg.extend([avg_glu, avg_asaz])
        sublist_sem.extend([sem_glu, sem_asaz])
    avg_free_phage.append(sublist_avg)
    sem_free_phage.append(sublist_sem)

# Append a placeholder for the 6th (empty) subplot (6 zeros)
avg_free_phage.append([0, 0, 0, 0, 0, 0])
sem_free_phage.append([0, 0, 0, 0, 0, 0])
```

```python
# X positions for the bars, with more space between pairs of experiments
x_pos = [1, 1.4, 3, 3.4, 5, 5.4]  # Custom x positions for 3 pairs of bars

# Color scheme for GLU and As/Az
color_scheme = ['#F0E68C', '#89CFF0']  # Light yellow for GLU, light blue for As/Az

# Phage labels for each subplot
phage_labels = ['λ', 'φ80', 'm13', 'T6', 'T5', '']

# Updated condition labels for the x-axis
condition_labels = ['Permissive', 'Resistant', 'Buffer']
```

```python
# Create a figure with 6 subplots (2 columns x 3 rows)
fig, axs = plt.subplots(3, 2, figsize=(12, 14))  # Adjust size as needed

# Loop over each subplot to create the bar charts
for i, ax in enumerate(axs.flatten()):
    if i == 5:
        ax.axis('off')  # Leave the last subplot blank
    else:
        avg_counts = avg_free_phage[i]  # Average free phage counts for current phage (6 values)
        sem_values = sem_free_phage[i]  # Corresponding SEM values (6 values)
        
        # Plot 3 pairs of bars (each pair: GLU and As/Az)
        for j in range(3):
            # GLU bar
            ax.bar(x_pos[2*j], avg_counts[2*j], color=color_scheme[0], width=0.3,
                   label='GLU' if j==0 else "", yerr=sem_values[2*j], capsize=5)
            # As/Az bar
            ax.bar(x_pos[2*j+1], avg_counts[2*j+1], color=color_scheme[1], width=0.3,
                   label='As/Az' if j==0 else "", yerr=sem_values[2*j+1], capsize=5)
        
        # Customize x-ticks and labels
        ax.set_xticks([1.2, 3.2, 5.2])
        ax.set_xticklabels(condition_labels, fontsize=16)
        ax.tick_params(axis='y', labelsize=16)
        
        # Set y-axis limit (with some extra headroom)
        ax.set_ylim(0, max(avg_counts) + 0.3*max(avg_counts))
        
        # Add axis labels and title
        ax.set_xlabel('Bacteria', fontsize=16)
        ax.set_ylabel('Phage particles/μl', fontsize=16)
        ax.set_title(f'phage {phage_labels[i]}', fontsize=18, fontweight='bold')
        
        
        ax.legend(loc='upper left', fontsize=14, frameon=False)

# Adjust layout and display
plt.tight_layout()

# Save the plot as an image
#plt.savefig("avg phage counts.png", dpi=300, bbox_inches='tight', facecolor='white')
#plt.savefig("avg phage counts.pdf", dpi=600, bbox_inches='tight', facecolor='white')
```

# Ratio =
### ((Free phages in As/Az)/(Free phages in buffer))/((Free phages in GLU)/(Free phages in buffer))

```python
# Compute ratios for each phage
ratio_lambda = R_lambda_AsAz / R_lambda_GLU
ratio_phi80  = R_phi80_AsAz  / R_phi80_GLU
ratio_m13    = R_m13_AsAz    / R_m13_GLU
ratio_T6     = R_T6_AsAz     / R_T6_GLU
ratio_T5     = R_T5_AsAz     / R_T5_GLU

# Flatten the ratio arrays to create a list for each phage,
# organized as in your example (pairs of R_P and R_R values)
bar_heights_lambda = ratio_lambda.flatten().tolist()
bar_heights_phi80  = ratio_phi80.flatten().tolist()
bar_heights_m13    = ratio_m13.flatten().tolist()
bar_heights_T6     = ratio_T6.flatten().tolist()
bar_heights_T5     = ratio_T5.flatten().tolist()

# Assemble the final bar_heights_list (last subplot is zeros)
bar_heights_list = [
    bar_heights_lambda,
    bar_heights_phi80,
    bar_heights_m13,
    bar_heights_T6,
    bar_heights_T5,
    [0, 0, 0, 0, 0, 0]  # Placeholder for subplot 6
]

#print("bar_heights_list = \n", bar_heights_list)
# Round each element in the bar_heights_list to three decimals
bar_heights_list_rounded = [[round(val, 3) for val in phage_list] for phage_list in bar_heights_list]

print("bar_heights_list =\n", bar_heights_list_rounded)


# --- Print average ratios and SEM for each phage ---
def print_ratio_stats(ratio_array, phage_name):
    # Assuming the ratios are stored in an array with the same shape as original:
    # Indices 0, 2, 4 correspond to R_P values and 1, 3, 5 to R_R values.
    flat = ratio_array.flatten()
    avg_R_P = np.mean(flat[[0, 2, 4]])
    sem_R_P = np.std(flat[[0, 2, 4]], ddof=1) / np.sqrt(3)
    avg_R_R = np.mean(flat[[1, 3, 5]])
    sem_R_R = np.std(flat[[1, 3, 5]], ddof=1) / np.sqrt(3)
    print(f"{phage_name} ratios:")
    print(f"  avg R_P ratio = {avg_R_P:.3f}, SEM = {sem_R_P:.3f}")
    print(f"  avg R_R ratio = {avg_R_R:.3f}, SEM = {sem_R_R:.3f}\n")

# Use the function for each phage
print_ratio_stats(ratio_lambda, "Lambda")
print_ratio_stats(ratio_phi80,  "φ80")
print_ratio_stats(ratio_m13,    "m13")
print_ratio_stats(ratio_T6,     "T6")
print_ratio_stats(ratio_T5,     "T5")
```

## Absorption rate ratio

 ## $\frac{\eta '}{\eta} = \frac{\mathrm{\langle lnR_{P, AsAz}\rangle}}{\mathrm{\langle lnR_{P, GLU}\rangle}}$

### Error Propagation for $ \eta'/\eta $


$$
\sigma_{\frac{\eta'}{\eta}} = \frac{\langle \ln R_{P, \text{AsAz}} \rangle}{\langle \ln R_{P, \text{GLU}} \langle} 
\sqrt{
\left(\frac{\mathrm{SEM}(\ln R_{P, \text{AsAz}})}{\langle \ln R_{P, \text{AsAz}} \rangle} \right)^2 + 
\left(\frac{\mathrm{SEM}(\ln R_{P, \text{GLU}})}{\langle \ln R_{P, \text{GLU}} \rangle} \right)^2
}
$$

```python
#  Compute η'/η and its error (error propagation) 

# Compute the ratio of means
eta_ratio_lambda = avg_X_lambda_AsAz / avg_X_lambda
eta_ratio_phi80 = avg_X_phi80_AsAz / avg_X_phi80
eta_ratio_m13 = avg_X_m13_AsAz / avg_X_m13
eta_ratio_T6 = avg_X_T6_AsAz / avg_X_T6
eta_ratio_T5 = avg_X_T5_AsAz / avg_X_T5

# Error propagation for η'/η:
# σ(η'/η) = (η'/η) * sqrt((σ_η'/η')² + (σ_η/η)²)
sem_eta_ratio_lambda = eta_ratio_lambda * np.sqrt((sem_X_lambda_AsAz / avg_X_lambda_AsAz)**2 + (sem_X_lambda / avg_X_lambda)**2)
sem_eta_ratio_phi80 = eta_ratio_phi80 * np.sqrt((sem_X_phi80_AsAz / avg_X_phi80_AsAz)**2 + (sem_X_phi80 / avg_X_phi80)**2)
sem_eta_ratio_m13 = eta_ratio_m13 * np.sqrt((sem_X_m13_AsAz / avg_X_m13_AsAz)**2 + (sem_X_m13 / avg_X_m13)**2)
sem_eta_ratio_T6 = eta_ratio_T6 * np.sqrt((sem_X_T6_AsAz / avg_X_T6_AsAz)**2 + (sem_X_T6 / avg_X_T6)**2)
sem_eta_ratio_T5 = eta_ratio_T5 * np.sqrt((sem_X_T5_AsAz / avg_X_T5_AsAz)**2 + (sem_X_T5 / avg_X_T5)**2)

# Store results in a DataFrame
eta_ratio_data = {
    "Phage": ["λ", "φ80", "m13", "T6", "T5"],
    "η'/η": [eta_ratio_lambda, eta_ratio_phi80, eta_ratio_m13, eta_ratio_T6, eta_ratio_T5],
    "δη": [sem_eta_ratio_lambda, sem_eta_ratio_phi80, sem_eta_ratio_m13, sem_eta_ratio_T6, sem_eta_ratio_T5]
}

df_eta_ratio = pd.DataFrame(eta_ratio_data)

# Display results
print(df_eta_ratio.to_string(index=False, float_format="{:.2f}".format))
```

# Calculate η 

```python
# Given values
B = 3e9                    # Bacterial density: 3 * 10^9
error_B = 0.5e9            # Error in B: 0.5 * 10^9
t_values = np.array([10, 30, 30, 55, 20])  # Incubation times (min) for each phage
error_t = 0.5              # Error in t (min)

# For each phage, assume the average X (from GLU condition) and its SEM have been computed:
# (Replace these with your actual computed values.)
# Example variable names from your previous code:
# avg_X_lambda, sem_X_lambda, avg_X_phi80, sem_X_phi80, avg_X_m13, sem_X_m13, avg_X_T6, sem_X_T6, avg_X_T5, sem_X_T5
avg_X = np.array([avg_X_lambda, avg_X_phi80, avg_X_m13, avg_X_T6, avg_X_T5])
sem_X = np.array([sem_X_lambda, sem_X_phi80, sem_X_m13, sem_X_T6, sem_X_T5])

# Compute eta for each phage: eta = avg_X / (B * t)
eta = avg_X / (B * t_values)

# Compute the relative errors for each term:
# Relative error in avg_X:
rel_err_X = sem_X / avg_X
# Relative error in B (constant for all phages):
rel_err_B = error_B / B
# Relative error in t:
rel_err_t = error_t / t_values

# Total relative error in eta:
rel_err_eta = np.sqrt(rel_err_X**2 + rel_err_B**2 + rel_err_t**2)

# Absolute error in eta:
eta_error = eta * rel_err_eta

# Build a DataFrame to display the results for each phage:
phages = ['λ', 'φ80', 'm13', 'T6', 'T5']
data_eta = {
    'Phage': phages,
    'Time (min)': t_values,
    'avg_X': avg_X,
    'SEM_X': sem_X,
    'eta': eta,
    'eta_error': eta_error
}

df_eta = pd.DataFrame(data_eta)
print(df_eta.to_string(index=False, float_format="{:.2e}".format))
df_eta
```

```python
from scipy.odr import ODR, Model, RealData
from scipy.stats import linregress

# ----------------------------------------------------------------
# Define color schemes for each phage:
colors = ['#0072B2', '#D55E00', '#009E73', '#CC79A7', '#E69F00']
# ----------------------------------------------------------------

# Define the linear model for ODR: f(x) = A * x + B,
# where x = log10(η)
def odr_linear(B, x):
    return B[0] * x + B[1]

# For fitting, exclude the last phage (T5)
filtered_eta = eta[:-1]
filtered_eta_ratio = eta_ratio[:-1]
filtered_log_eta = np.log10(filtered_eta)

# Compute errors in log_eta from the errors in η:
# σ_log_eta = σ_η / (η * ln(10))
filtered_eta_error = eta_error[:-1]  # x uncertainties for filtered data
log_eta_error = filtered_eta_error / (filtered_eta * np.log(10))

# Use the provided y uncertainties for filtered data
filtered_eta_ratio_error = eta_ratio_error[:-1]

# Create a RealData object for ODR, using x = log10(η)
data = RealData(filtered_log_eta, filtered_eta_ratio, sx=log_eta_error, sy=filtered_eta_ratio_error)

# Create a Model object for the linear model
linear_odr_model = Model(odr_linear)

# Set up ODR with an initial guess for [A, B]
odr_instance = ODR(data, linear_odr_model, beta0=[1, 0])
output = odr_instance.run()

# Extract fitted parameters
A, B = output.beta

# Generate fit-line data:
# x_fit is defined in the original η space, but the model uses log10(η)
x_fit = np.logspace(np.log10(min(eta)) - 0.1, np.log10(max(eta)) + 0.1, 100)
y_fit = A * np.log10(x_fit) + B

# Compute Pearson's r for the filtered fit (using log10(η))
slope, intercept, r_value, p_value, std_err = linregress(filtered_log_eta, filtered_eta_ratio)
pearson_r = round(r_value, 2)

# Reduced chi-square from ODR (res_var is the residual variance)
reduced_chi2 = output.res_var

# ---- Plotting ----
fig, ax = plt.subplots(figsize=(10, 7))
for i in range(len(eta)):
    ax.errorbar(eta[i], eta_ratio[i],
                xerr=eta_error[i], yerr=eta_ratio_error[i],
                fmt='o', color=colors[i], markersize=10,
                capsize=5, markeredgecolor='black', zorder=3)
    # Print phage labels near the data points
    ax.text(eta[i], eta_ratio[i] + 0.04, phages[i], ha='center', fontsize=20)

# Plot the fitted line from the ODR fit
ax.plot(x_fit, y_fit, color='black', linestyle='--', linewidth=2,
        label=(f'ODR Fit: η\'/η = {A:.1f} log(η) + {B:.1f}\n'
               f'Pearson r = {pearson_r:.2f}\n'
               f'$\\mathrm{{\\chi^2_{{ODR}}}}$ = {reduced_chi2:.2f}'))

# Customize the plot
ax.set_xscale('log')
ax.set_xlim(min(eta)*0.8, max(eta)*1.2)
ax.set_ylim(0, max(eta_ratio)*1.2)
ax.set_xlabel("η", fontsize=22)
ax.set_ylabel("η'/η", fontsize=22)
ax.tick_params(axis='both', which='major', labelsize=22)
ax.legend(fontsize=20, loc='upper left')

plt.tight_layout()
#plt.savefig("eta_ratio_vs_eta.png", dpi=300, bbox_inches='tight', facecolor='white')
#plt.savefig("eta_ratio_vs_eta.pdf", dpi=600, bbox_inches='tight', facecolor='white')
plt.show()
```
